To make the assistant ready for multiple questions, we can extend the example you provided by allowing dynamic user input.

```python
import openai

# Function to interact with the fine-tuned model
def ask_question(question):
    response = openai.ChatCompletion.create(
        model=model_id,  # Your fine-tuned model ID
        messages=[
            {"role": "system", "content": "You are an AI lawyer specializing in Indian law. Answer queries with accurate legal advice, explanations in simple language, and a cool slangily tone."},
            {"role": "user", "content": question}
        ]
    )
    return response['choices'][0]['message']['content']

# Testing the interaction
question = "I received a speeding ticket, what are my options?"
response = ask_question(question)
print("AI Legal Assistant Response:", response)
```
